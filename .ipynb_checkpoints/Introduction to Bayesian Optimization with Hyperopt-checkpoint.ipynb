# Objective

```python
x = np.linspace(-5,5,7)
y = [2, -1.5, 4, 1, -5, 1.5, 12]

plt.plot(x,y,'bo', label = 'obs')
z = np.polyfit(x, y, 6)
E = np.poly1d(z)

x_lin = np.linspace(-6, 5, 100)
f_x = E(x_lin)

plt.plot(x_lin, f_x, color = 'green', label = 'f(x)');
```

```python
def poly(x):
    f = np.poly1d([1, -2, -26, 28, 145, -26, -80])

    return 0.01 * f(x)
```

```python
x = np.linspace(-4, 5, 10000)
y = poly(x)
plt.plot(x, y);

print('Minimum of %0.4f occurs at %0.4f' % (min(y), x[np.argmin(y)]))
```
